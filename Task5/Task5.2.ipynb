```python
import numpy as np
```

```python
k = 3
T = 300
Episodes = 100
```

```python
def uniform_initialization(mu_range=(-1, 1), sigma_range=(0.2, 1.2)):
    mus = np.random.uniform(mu_range[0], mu_range[1], k)
    sigmas = np.random.uniform(sigma_range[0], sigma_range[1], k)
    return mus, sigmas

def run_experiments(strategies, T=300):
    results = {strategy: [] for strategy in strategies}
    for strategy in strategies:
        for e in range(Episodes):
            error = run_prediction_bandit(strategy, e+1)
            results[strategy].append(error)
    return results

def run_prediction_bandit(strategy, e, T1=100):
    mus, sigmas = uniform_initialization()
    predictions = []
    errors = []
    actions = []
    observations = {i: [] for i in range(k)}

    for t in range(T):
        if strategy == 'random':
            if t < T1:
                action = (t % 3)
            else:
                min_std = 10
                min_std_action = 0
                for action in range(k):
                    if min_std > np.std(observations[action]):
                        min_std = np.std(observations[action])
                        min_std_action = action
                action = min_std_action
            prediction = np.mean(observations[action]) if observations[action] else 0
        else:
            print("Not yet implemented")
        observation = np.random.normal(mus[action], sigmas[action])
        error = abs(observation - prediction)
        predictions.append(prediction)
        errors.append(error)
        actions.append(action)
        observations[action].append(observation)
    print(f"Error for episode {e} : {sum(errors)}")
    print('-'*60)
    return sum(errors)

```

```python
strategies = ['random']
experiment_results = run_experiments(strategies)
print("Mean result: ", np.mean(experiment_results['random']))
```
